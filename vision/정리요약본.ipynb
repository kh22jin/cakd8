## 반복문 while

```
for문은 반복횟수가 정해져 있을 때 사용
while문은 반복횟수가 정해져 있지 않을 때 사용

```

````python
import time
i = 0
while True:
    print(i, end = ", ")
    i += 1
    time.sleep(1)    # 무한루프 # 1초마다 나타내준다.
    if i == 10:
        break     # 끝내준다.
print("끝")
````

```
time.sleep(a) : a초마다 나타내준다, 무한루프
break : time.sleep을 멈춰준다.
continue : 넘어가라는 뜻, 제어흐름(반복)을 유지한 상태에서 코드의 실행만 건너뛰는 역할
```

````python
# 3의 배수가 아닌 것만 출력해라.
i = 0
while i<10:
    i += 1
    if i%3==0:
        continue
    print(i)
````

## 클래스 만들기

````python
class FourCal:
    def setdata(self, a, b):
        pass

class FourCal:
    def setdata(self, a, b):
        self.first = a
        self.second = b   #설계도

cal = FourCal()     #객체

cal.setdata(1, 2)   #동사

cal.first          #형용사
````

```
객체 : object, instance, 오브젝트, 인스턴스
     모든 객체는 두가지를 가지고 있다.
     동사: 메소드, method(함수)
     형용사: 속성, 애트리뷰트, attribute
```

````python
class FourCal:
    def setdata(self, a, b):
        self.first = a
        self.second = b   # setdata는 먼저 반드시 시행이 되어야한다.
    def add(self):          # setdata로 값을 먼저 줘야 add, subtract가 작동한다.
        self.result = self.first + self.second
        return self.result
    def subtract(self):
        self.result = self.first - self.second
        return self.result
````

## 초기화 함수, 생성자

````python
class FourCal:
    # 만들지 않아도 실행되는 함수
    def __init__(self):
        pass
    # 초기화, 생성자(constructor)
    # 초기화해주는 함수, 무조건 실행되는 함수
    def __init__(self, a, b):
        self.first = a
        self.second = b   
    def add(self):
        self.result = self.first + self.second
        return self.result
    def subtract(self):
        self.result = self.first - self.second
        return self.result
````

````python
 #부모클래스

class FourCal:           
    def __init__(self, a, b):
        self.first = a
        self.second = b   
    def add(self):
        self.result = self.first + self.second
        return self.result
    def subtract(self):
        self.result = self.first - self.second
        return self.result
    def multiply(self):
        self.result = self.first * self.second
        return self.result
    def divide(self):
        self.result = self.first / self.second
        return self.result
````

````python
 # 자식클래스

class Child(FourCal):   
    pass
````

````python
#상속받은 함수(부모클래스)에 필요한 게 없으면 만들어서 사용 가능
class Child(FourCal):
    def power(self):     #거듭제곱함수
        self.result = self.first ** self.second
        return self.result
````

## 메소드 오버라이딩

overriding 메소드 재정의

````python
# 메소드를 새로 만들어 부모클래스에 덮어쓸 수 있다.

class Child(FourCal):
    def power(self): 
        self.result = self.first ** self.second
        return self.result
    def divide(self):
        if self.second != 0:   # b에 0이 들어갈 때 나누기는 에러가 뜨기 때문에 조건을 추가한다.
            self.result = self.first / self.second
            return self.result
        else:
            print("0으로 나누셨군요")  
````

## threading 
: 스레드를 다루는 모듈

- 프로세스 : 컴퓨터에서 동작하고 있는 프로그램들

- 스레드 : 보통은 1개의 프로세스가 한 가지 일만 하지만 스레드를 사용하면 프로세스 안에 프로세스를 쪼개서 2가지 또는 그 이상의 일을 동시에 수행할 수 있다.

````python
#  long_task 함수는 수행하는 데 5초의 시간이 걸리는 함수, 이 함수를 5번 반복하여 총 25초의 시간이 걸린다.
import time
def long_task():
    for i in range(5):
        time.sleep(1)
        print("working: %s"  %i)
print("Start")
for _ in range(5):
    long_task()
print("End")
````

````python
# threading으로 수정하여 작업 시간을 단축하게 만들어주기

import time
import threading

def long_task():
    for i in range(5):
        time.sleep(1)
        print("working: %s"  %i)
print("Start")

threads = []

for _ in range(5):
    t = threading.Thread(target = long_task)
    threads.append(t)       # 스레드를 5개 생성한다.
    
for t in threads:
    t.start()       # 스레드가 동시에 실행된다.
for t in threads:
    t.join()         # 스레드가 종료될 때까지 기다렸다가 밑에 End가 실행된다.
    
print("End")
````

## 메타문자

(특수문자 중에서 의미를 부여한 특수문자)

### [ ]
: [ ]사이의 문자들과 매치

```
[a-zA-Z] : 모든 알파벳 = [a-Z]
[0-9] : 숫자 = \d

\d : 숫자와 매치, [0-9]와 동일한 표현식
\D : 숫자가 아닌 것과 매치, [^0-9]와 동일한 표현식

\s : space나 tab처럼(whitespace) 공백을 표현하는 문자[\t\n\r\f\v] 와 동일한 표현식, 
맨 앞의 빈 칸은 공백문자를 의미한다.
\S : [^ \t\n\r\f\v]

\w : 문자+숫자 와 매치 [a-zA-Z0-9_] 
\W : 문자+숫자 가 아닌 문자와 매치 [^a-zA-Z0-9_]

# ^ : 대괄호 안에 있을 때만 반대 (not) 라는 의미
```

: 줄바꿈 문자인 \n을 제외한 모든 문자

```
ex) 

a.b : aab, a0b, ...
a[.]b : 눈에 보이는 .(점) 그대로를 의미한다.
```

```
ex) ca*t : 
ct, cat, caaat, ...
```

### 반복 (+)
: 최소 1번 이상 반복될 때 사용한다.

````python
# 방법 1 search를 사용

data = ["010-1234-5678", "010-1234-5679", "010-1234-5670", "010-1234-670", "02-334-2243"] 


def is_phone_number(data):
    phone_number = []
    for i in data:
        if re.search("\d{3}-\d{4}-\d{4}", i):   
            phone_number.append(i)
    return phone_number


is_phone_number(data)
````

````python
#방법 2 finditer를 사용

data = '"010-1234-5678", "010-1234-5679", "010-1234-5670", "010-1234-670", "02-334-2243"'

def is_phone_number(data):
    phone_number = []
    for i in re.finditer("\d{3}-\d{4}-\d{4}", data):         
        phone_number.append(i.group())
    return phone_number

is_phone_number(data)
````

````python
# 이미지를 전치시켜서 출력하세요

with open("lena.raw", "rb") as f:
    img = np.fromfile(f, dtype = 'ubyte', sep = "")
    img = np.reshape(img, [512,512])
    img = img.T
    
plt.imshow(img, cmap = 'gray')
````

````python
# 이미지를 반전시켜보세요.

with open("lena.raw", "rb") as f:
    img = np.fromfile(f, dtype = 'ubyte', sep = "")
    img = np.reshape(img, [512,512])
    img = 255 - img
    
plt.imshow(img, cmap = 'gray')
````

## 피벗 테이블(평균)
: 피벗테이블의 기본 연산은 평균을 구해준다.

````python
tips = pydataset.data('tips')
tips['tip_pct'] = tips['tip'] / tips['total_bill'] 
tips[:2]
````

````python
tips.groupby(['day', 'smoker']).agg('mean')   # 묶어서 계산하는법
````

````python
tips.pivot_table(values = 'tip_pct', index = 'time', columns = 'smoker')     # 평균 구해준다.
````

````python
tips.pivot_table(['tip_pct', 'size'], index = ['time', 'day'],          # 여러 개로도 가능하다.
                 columns = 'smoker')
````

### 중간 평균 확인

````python
# 피벗테이블은 중간평균을 구할 수 있다.

tips.pivot_table(values = 'tip_pct', index = 'time', columns = 'smoker',
                margins = True)
````

### 다른 함수로 사용

````python
# 평균 말고 다른 함수를 사용하려면 aggfunc를 넣어주면 된다. (count, len, sum, ...)

tips.pivot_table(values = 'tip_pct', index = ['time', 'day'], columns = 'smoker',
                margins = True, aggfunc = 'count')
````

### 리샘플링
: 시계열의 빈도를 변환하는 과정

#### 업샘플링 
: 하위 빈도의 데이터를 상위 빈도로 집계하는 것

````python
ts = pd.Series(np.random.randn(100), 
               index = pd.date_range('2000-01-01', periods = 100, freq = 'D'))
ts
````

````python
ts.resample('W').mean()
````

````python
ts.resample('M').mean()
````

````python
ts.resample('M', kind = 'period').mean()    #일 빼고 월별로만 보여준다
````

#### 다운샘플링 
: 상위 빈도의 데이터를 하위 빈도로 집계하는 것

````python
rng = pd.date_range('2000-01-01', periods = 12, freq = 'T')     #분 단위로 쪼개져있다.

ts = pd.Series(np.arange(12), index = rng)
ts
````

````python
list(ts.resample('5min'))
````

````python
ts.resample('5min').sum()       #5분단위로 쪼개준다.
````

````python
ts.resample('5min', closed = 'right').sum()                      
 # closed = ' '     :      ' '는 닫혀있다, 포함한다.
````

- OHLC 리샘플링 : 시가, 고가, 저가, 종가의 값을 계산해준다.

````python
ts.resample('5min').ohlc()
````
